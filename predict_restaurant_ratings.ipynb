```python
# Load dataset
df = pd.read_csv('/content/Dataset .csv')
df
```

```python
# Remove irrelevant columns
drop_cols = ['Restaurant ID', 'Restaurant Name', 'Address', 'Locality', 'Locality Verbose', 'Rating text', 'Rating color', 'Currency', 'City']
df_filtered = df.drop(columns=drop_cols, errors='ignore')
df_filtered
```

```python
# Fill missing values
df_filtered['Cuisines'].fillna('NA', inplace=True)
df_filtered
```

```python
# Encode categorical features using Label Encoding
label_cols = ['Has Table booking', 'Has Online delivery', 'Is delivering now']
for column in label_cols:
    df_filtered[column] = LabelEncoder().fit_transform(df_filtered[column])
    print(df_filtered[column].value_counts())
```

```python
# Assign numerical features
num_features = ['Average Cost for two', 'Price range', 'Votes', 'Latitude', 'Longitude', 'Cuisines']
df_numerical = df_filtered[num_features + ['Aggregate rating']]
df_numerical
```

```python
# Assuming 'num_features' contains the names of your numerical columns
num_features = ['Average Cost for two', 'Price range', 'Votes', 'Latitude', 'Longitude']  # Removed 'Cuisines'

# Create a separate DataFrame with only the numerical features
numerical_data = df_filtered[num_features]

# Apply MinMaxScaler to the numerical DataFrame
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(numerical_data)

# Update the original DataFrame with the scaled numerical values
df_filtered[num_features] = scaled_data
df_filtered
```

```python
# Split dataset into training and testing sets
X = df_numerical.drop(columns=['Aggregate rating'])
y = df_numerical['Aggregate rating']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.50, random_state=21)
```
