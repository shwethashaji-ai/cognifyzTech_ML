```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Read the dataset
data = pd.read_csv("/content/Dataset .csv")
data
```

```python
data.head(5)
```

```python
# Handle missing values using forward fill
data.fillna(method='ffill', inplace=True)
data
```

```python
print(data.columns)
print(data.head())
```

```python
# Encode the target variable 'Cuisines'
encoder = LabelEncoder()
data['Cuisines'] = encoder.fit_transform(data['Cuisines'])
```

```python
# Selecting relevant features for model training
selected_features = ['Price range', 'Aggregate rating', 'Votes']  # Modify as needed
feature_data = data[selected_features]
feature_data
```

```python
# Splitting data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(feature_data, data['Cuisines'], test_size=0.2, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
# Initialize and train a random forest classifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
y_pred
```

```python
# Evaluate model performance
print(f'Accuracy Score: {accuracy_score(y_test, y_pred):.2f}')

```

```python
print(classification_report(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
# Display the mapping of cuisines
cuisine_list = dict(zip(encoder.classes_, range(len(encoder.classes_))))
print("Cuisine Mapping:", cuisine_list)
```

```python

# Create a DataFrame for the mapping
cuisine_mapping_df = pd.DataFrame({'Cuisine': encoder.classes_, 'Numerical Label': range(len(encoder.classes_))})

# Display the DataFrame
print(cuisine_mapping_df)
```

```python

# Print the mapping with formatting
for cuisine, label in cuisine_list.items():
    print(f"{cuisine}: {label}")
```
